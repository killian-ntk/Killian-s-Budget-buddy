### Foundation Data Processing Functions

````python
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        file_path: Path to CSV file or file object
    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    if isinstance(file_path, pd.DataFrame):
        df = file_path.copy()
    else:
        df = pd.read_csv(file_path)

    # Validate required columns
    required_columns = ['Date', 'Amount', 'Category', 'Description']
    missing_columns = [col for col in required_columns if col not in df.columns]
    if missing_columns:
        raise ValueError(f"❌ Missing required columns: {missing_columns}")

    # Clean Amount column - remove dollar signs and convert to numeric
    df['Amount'] = df['Amount'].astype(str).str.replace('$', '', regex=False)
    df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')

    # Handle missing values
    df = df.dropna(subset=['Amount'])
    df['Category'] = df['Category'].fillna('Uncategorised')
    df['Description'] = df['Description'].fillna('No description')

    # Separate spending from refunds
    df['Type'] = df['Amount'].apply(lambda x: 'Refund' if x < 0 else 'Spending')

    print(f"✅ Data loaded successfully! {len(df)} transactions found.")
    return df

# Test your function
clean_data = load_and_clean_transaction_data(df_sample)
print(clean_data)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df: Cleaned transaction DataFrame
    Returns:
        dict: Analysis results and insights
    """
    # Separate spending and refunds for accurate analysis
    spending_df = df[df['Type'] == 'Spending']
    refunds_df = df[df['Type'] == 'Refund']

    # Calculate total spending and refund amounts
    total_spending = spending_df['Amount'].sum()
    total_refunds = abs(refunds_df['Amount'].sum())
    net_spending = total_spending - total_refunds

    # Group transactions by category and calculate business metrics
    category_totals = spending_df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
    category_counts = spending_df.groupby('Category')['Amount'].count()
    category_averages = spending_df.groupby('Category')['Amount'].mean()

    # Calculate percentage of total spending per category
    category_percentages = (category_totals / total_spending * 100).round(2)

    # Identify top 3 highest spending categories
    top_3_categories = category_totals.head(3)

    # Identify largest single purchase
    largest_purchase_idx = spending_df['Amount'].idxmax()
    largest_purchase = spending_df.loc[largest_purchase_idx]

    # Flag categories exceeding 30% of total budget
    high_spend_categories = category_percentages[category_percentages > 30].index.tolist()

    # Print business-friendly summary
    print("=" * 50)
    print("📊 BUDGET BUDDY - SPENDING ANALYSIS REPORT")
    print("=" * 50)
    print(f"\n💰 Total Spending:   ${total_spending:.2f}")
    print(f"↩️  Total Refunds:    ${total_refunds:.2f}")
    print(f"📉 Net Spending:     ${net_spending:.2f}")
    print(f"\n📂 Spending by Category:")
    for category in category_totals.index:
        print(f"   {category:<20} ${category_totals[category]:.2f}  ({category_percentages[category]:.1f}%)")
    print(f"\n🏆 Top 3 Spending Categories:")
    for i, category in enumerate(top_3_categories.index, 1):
        print(f"   {i}. {category}: ${top_3_categories[category]:.2f}")
    print(f"\n🔺 Largest Single Purchase: {largest_purchase['Description']} (${largest_purchase['Amount']:.2f})")
    if high_spend_categories:
        print(f"\n⚠️  High Spend Alert: {', '.join(high_spend_categories)} exceeds 30% of total budget")

    # Return results as dictionary for use in other functions
    return {
        'total_spending': total_spending,
        'total_refunds': total_refunds,
        'net_spending': net_spending,
        'category_totals': category_totals,
        'category_percentages': category_percentages,
        'category_averages': category_averages,
        'category_counts': category_counts,
        'top_3_categories': top_3_categories,
        'largest_purchase': largest_purchase,
        'high_spend_categories': high_spend_categories
    }

# Test your function
analysis = analyze_spending_patterns(clean_data)
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis

    🤖 AI Collaboration Prompt:
    "Based on spending analysis data, create professional financial
    recommendations. Include specific savings opportunities, spending
    pattern observations, and actionable advice formatted for a
    personal finance app user."

    Args:
        analysis_data: Dictionary with spending analysis results
    Returns:
        str: Formatted recommendations report
    """
    recommendations = []

    # Extract data from analysis dictionary
    total_spending = analysis_data['total_spending']
    total_refunds = analysis_data['total_refunds']
    net_spending = analysis_data['net_spending']
    category_totals = analysis_data['category_totals']
    category_percentages = analysis_data['category_percentages']
    top_3_categories = analysis_data['top_3_categories']
    high_spend_categories = analysis_data['high_spend_categories']
    largest_purchase = analysis_data['largest_purchase']

    # Build recommendations report header
    report = []
    report.append("=" * 50)
    report.append("💡 BUDGET BUDDY - FINANCIAL RECOMMENDATIONS")
    report.append("=" * 50)

    # Observation 1: Overall spending summary
    report.append(f"\n📋 SPENDING SUMMARY")
    report.append(f"   Your total spending this period was ${total_spending:.2f}.")
    if total_refunds > 0:
        report.append(f"   After ${total_refunds:.2f} in refunds, your net spending is ${net_spending:.2f}.")

    # Observation 2: Top spending category advice
    top_category = top_3_categories.index[0]
    top_amount = top_3_categories.iloc[0]
    top_percentage = category_percentages[top_category]
    report.append(f"\n🔍 TOP SPENDING INSIGHT")
    report.append(f"   Your biggest expense is {top_category} at ${top_amount:.2f} ({top_percentage:.1f}% of total).")

    # Generate specific advice based on top category
    category_advice = {
        'Dining Out': "Consider meal prepping at home a few days a week to reduce dining costs.",
        'Entertainment': "Look for free or discounted events in your area to cut entertainment costs.",
        'Coffee': "Making coffee at home could save you over $100 per month.",
        'Groceries': "Try shopping at discount supermarkets or buying in bulk to reduce grocery bills.",
        'Transport': "Consider a weekly or monthly transport pass for better value.",
        'Utilities': "Review your utility plans — switching providers could reduce your bills.",
        'Rent': "Your rent is your biggest fixed cost. Consider splitting costs with housemates if possible."
    }
    if top_category in category_advice:
        report.append(f"   💡 Tip: {category_advice[top_category]}")
    else:
        report.append(f"   💡 Tip: Review your {top_category} spending to find potential savings.")

    # Observation 3: High spend category warnings
    if high_spend_categories:
        report.append(f"\n⚠️  HIGH SPEND WARNINGS")
        for category in high_spend_categories:
            report.append(f"   {category} accounts for {category_percentages[category]:.1f}% of your budget — consider setting a spending limit.")

    # Observation 4: Largest purchase note
    report.append(f"\n🔺 LARGEST PURCHASE")
    report.append(f"   Your biggest single transaction was {largest_purchase['Description']} at ${largest_purchase['Amount']:.2f}.")
    if largest_purchase['Amount'] > total_spending * 0.25:
        report.append(f"   This single purchase made up more than 25% of your total spending.")

    # Observation 5: Savings opportunities
    report.append(f"\n✅ SAVINGS OPPORTUNITIES")
    wants_categories = ['Dining Out', 'Entertainment', 'Coffee']
    wants_total = sum(category_totals.get(cat, 0) for cat in wants_categories if cat in category_totals)
    if wants_total > 0:
        report.append(f"   You spent ${wants_total:.2f} on non-essential items (Dining Out, Entertainment, Coffee).")
        report.append(f"   Reducing these by 20% could save you ${wants_total * 0.2:.2f} this period.")

    report.append(f"\n{'=' * 50}")
    report.append("Keep tracking your spending with Budget Buddy! 💰")
    report.append("=" * 50)

    # Join and print the report
    final_report = "\n".join(report)
    print(final_report)
    return final_report

# Test your function
recommendations = generate_financial_recommendations(analysis)
print(recommendations)
````
